```python
import pandas as pd

data ={
    "email":[
        "Win money now",
        "Hello how are you",
        "Claim your free prize",
        "Meeting at 5pm",
        "Congratulation you won the lottery",
        "Lets's catch up today",
        "Free entry in contest",
        "Project discussion today"
    ],
    "label":[1,0,1,0,1,0,1,0]

}

df = pd.DataFrame(data)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
x = vectorizer.fit_transform(df["email"])
y = df["label"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier()
gb.fit(x_train, y_train)
```

```python
test_email = ["Lets's catch up today"]

test_vector = vectorizer.transform(test_email)

def explain(pred):
  return "spam" if pred == 1 else "not spam"

print("DT:", explain(dt.predict(test_vector)[0]))
print("RF:", explain(rf.predict(test_vector)[0]))
print("GB:", explain(gb.predict(test_vector)[0]))
```
